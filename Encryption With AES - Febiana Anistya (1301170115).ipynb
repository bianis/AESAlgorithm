# Initial Round + STEP 4 - AddRoundKey

```python
def add_round_key(A, k):
    for i in range(4):
        for j in range(4):
            A[i][j] ^= k[i][j]
```

# STEP 1 - SubBytes

```python
def sub_bytes(s):
    for i in range(4):
        for j in range(4):
            s[i][j] = s_box[s[i][j]]
```

# STEP 2 - ShiftRows

```python
def shift_rows(s):
    s[1][0], s[1][1], s[1][2], s[1][3] = s[1][1], s[1][2], s[1][3], s[1][0]
    s[2][0], s[2][1], s[2][2], s[2][3] = s[2][2], s[2][3], s[2][0], s[2][1]
    s[3][0], s[3][1], s[3][2], s[3][3] = s[3][3], s[3][0], s[3][1], s[3][2]
```

```python
def keySchedule(key, idRound):
    temp = []
    temp.append(key[1][3])
    temp.append(key[2][3])
    temp.append(key[3][3])
    temp.append(key[0][3])

    temp2 = []
    temp2.append(key[0][0])
    temp2.append(key[1][0])
    temp2.append(key[2][0])
    temp2.append(key[3][0])
    
    RCon = [[0x01, 0x00, 0x00, 0x00],
            [0x02, 0x00, 0x00, 0x00],
            [0x04, 0x00, 0x00, 0x00],
            [0x08, 0x00, 0x00, 0x00],
            [0x10, 0x00, 0x00, 0x00],
            [0x20, 0x00, 0x00, 0x00],
            [0x40, 0x00, 0x00, 0x00],
            [0x80, 0x00, 0x00, 0x00],
            [0x1b, 0x00, 0x00, 0x00],
            [0x36, 0x00, 0x00, 0x00]]

    for i in range(4):
        temp[i] = s_box[temp[i]]

    n = 0
    c1 = []
    for i in temp:
        c1.append(i ^ RCon[idRound][n] ^ temp2[n])
        n += 1
    
    c2 = []
    for e in range(4):
        c2.append(key[e][1] ^ c1[e])

    c3 = []
    for e in range(4):
        c3.append(key[e][2] ^ c2[e])

    c4 = []
    for e in range(4):
        c4.append(key[e][3] ^ c3[e])

    newkey = []
    for i in range(4):
        newkey.append([c1[i], c2[i], c3[i], c4[i]])

    c = 0    
    for i in range(4):
        for o in range(4):
            key[o][i] = newkey[o][i]
            c += 1
    return key
```

# PROSES ENKRIPSI

```python
def printM(s):
    print('\n'.join([''.join(['{:4}'.format(hex(item)[2:]) for item in row]) 
      for row in s]))
```

```python
def printbin(s):
    for r in s:
        print(r)
```

```python
import copy
print('ENKRIPSI AES')
A = [[0x73, 0x61, 0x79, 0x61],
     [0x73, 0x75, 0x6b, 0x61],
     [0x6d, 0x61, 0x69, 0x6e],
     [0x61, 0x6c, 0x67, 0x6f]]

K = [[0x62, 0x75, 0x61, 0x74],
     [0x6b, 0x65, 0x79, 0x73],
     [0x6c, 0x61, 0x67, 0x69],
     [0x64, 0x69, 0x72, 0x69]]

originalKey = copy.deepcopy(K)

expkey = []

print('============================')
print('Round 0')
print('============================')
printM(A)
print()
print('Add Round Key')
add_round_key(A, K)
printM(A)
print()

for i in range(9):
    print('============================')
    print('Round ',i+1)
    print('============================')
    print()
    print('R',i+1,'Step 1 - Sub Bytes')
    print('----------------------')
    sub_bytes(A)
    printM(A)
    print()
    shift_rows(A)
    print('R',i+1,'Step 2 - Shift Rows')
    print('----------------------')
    printM(A)
    print()

    print('R',i+1,'Step 3 - Mix Column')
    print('----------------------')
    mixColumn(A)
    printM(A)
    print()
    K = keySchedule(K, i)
    expkey.append(copy.deepcopy(K))
    print('R',i+1,'Step 4 - Add Round Key')
    print('----------------------')
    print('Round Key', i+1,':')
    printM(K)
    print()
    add_round_key(A, K)
    print('Hasil:')
    printM(A)
    print()
    
print('============================')
print('Round 10')
print('============================')
printM(A)
print()
print('R 10 Step 1 - Sub Bytes')
print('----------------------')
sub_bytes(A)
printM(A)
print()
print('R 10 Step 2 - Shift Rows')
print('----------------------')
shift_rows(A)
printM(A)
print()
print('R 10 Step 4 - Add Round Key')
print('----------------------')
K = keySchedule(K,9)
expkey.append(copy.deepcopy(K))
print('Round Key 10:')
printM(K)
print()
add_round_key(A, K)
print('Hasil:')
printM(A)
print()
```
